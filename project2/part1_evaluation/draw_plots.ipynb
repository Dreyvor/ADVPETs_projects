```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.rcdefaults()

labels = ['GEN_KEYS', 'ISSUANCE', 'SHOW_CREDS', 'VERIFY_SIGNED_REQUEST', 'GLOBAL']
means = np.array([5.86214, 17.53190, 4.22974, 15.32128, 43.03176])
std = np.array([0.24969, 0.73028, 0.21414, 0.57945, 1.61744])
width = 0.35       # the width of the bars: can also be len(x) sequence
yticks = np.arange(len(labels))

fig, ax = plt.subplots(figsize=(10,3))

ax.barh(yticks, means, width, xerr=std, align='center', tick_label=labels)
ax.invert_yaxis()
ax.set_xlabel("Time [ms]")
ax.set_title('Time by operations (1 user doing 1 query 10000 times for avg and std)')
ax.grid(alpha=.35)

for i, (m,s) in enumerate(zip(means,std)):
    ax.text(m + 1, i-.1, str(m)+" +/- "+str(s))

plt.xlim([0,55])
plt.show()
```
