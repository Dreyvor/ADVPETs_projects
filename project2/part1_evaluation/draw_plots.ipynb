```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.rcdefaults()

labels = ['GEN_KEYS', 'ISSUANCE', 'SHOW_CREDS', 'VERIFY_SIGNED_REQUEST', 'GLOBAL']
means = np.array([round(v, 2) for v in [5.86214, 17.53190, 4.22974, 15.32128, 43.03176]])
std = np.array([round(v, 2) for v in [0.24969, 0.73028, 0.21414, 0.57945, 1.61744]])
width = 0.35       # the width of the bars: can also be len(x) sequence
xticks = np.arange(len(labels))

fig, ax = plt.subplots(figsize=(9.5,5))

ax.bar(xticks, means, width, yerr=std, align='center', tick_label=labels)
ax.set_ylabel("Time [ms]")
ax.set_title('Time by operations (1 user doing 1 query 10000 times for avg and std)')
ax.grid(alpha=.35)

for i, (m,s) in enumerate(zip(means,std)):
    ax.text(i-.1, m + 1, str(m)+" +/- "+str(s), rotation='vertical')

plt.ylim([0,60])
plt.show()
```
